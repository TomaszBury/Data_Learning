# The .between() Method

```python
import pandas as pd
employees = pd.read_csv('employees.csv', parse_dates = ['Start Date', 'Last Login Time'])
employees['Senior Management'] = employees['Senior Management'].astype('bool')
employees['Gender'] = employees['Gender'].astype('category')
employees.head(n=2)
```

```python
mask_salary_range = employees['Salary'].between(60000, 70000)
employees[mask_salary_range]
```

```python
mask_bonus = employees['Bonus %'].between(2.0, 5.0)
employees[mask_bonus]
```

```python
mask_start_date = employees['Start Date'].between('1990-01-01', '1999-12-31')
employees[mask_start_date].sort_values(by='Start Date')
```

```python
mask_start_date = employees['Start Date'].between('1991-01-01', '1992-10-01')
employees[mask_start_date].sort_values(by=['Start Date'])
```

```python
mask_login = employees['Last Login Time'].between('8:30AM', '12:00PM')
employees[mask_login].sort_values(by='Last Login Time', ascending=False)
```
