# The .duplicated() Method

```python
import pandas as pd
employees = pd.read_csv('employees.csv', parse_dates = ['Start Date', 'Last Login Time'])
employees['Senior management'] = employees['Senior Management'].astype('bool')
employees['Gender'] = employees['Gender'].astype('category')
employees.head(n=2)
```

```python
employees = employees.sort_values(by='First Name')
```

```python
employees.head(n=4)
```

```python
employees['First Name'].duplicated().head(n=5)
```

```python
mask_first_name = employees['First Name'].duplicated()
employees[mask_first_name].head(n=6)
```

```python
mask_first_name = employees['First Name'].duplicated(keep='last')
employees[mask_first_name].head(n=6)
```

```python
employees['First Name'].head(n=6)
```

```python
mask_first_name = employees['First Name'].duplicated(keep = False)
employees[mask_first_name].head(n=6)
```

```python
mask_first_name = ~employees['First Name'].duplicated(keep = False)
employees[mask_first_name]
```

```python
employees.drop_duplicates(subset=['Team'], keep=False)
```

```python
employees.drop_duplicates(subset=['First Name', 'Team']).head(n=4)
```
