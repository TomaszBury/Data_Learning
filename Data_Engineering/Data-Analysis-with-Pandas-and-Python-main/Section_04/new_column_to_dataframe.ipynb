# Create New Column from existing column

```python
nba = pd.read_csv("nba.csv")
nba.head(2)
```

```python
nba['Age'].head(7)
```

```python
nba['Age'].add(2).head(7)
```

```python
nba['Age in a Decade'] = nba['Age'].add(10)
nba.head(7)
```

```python
nba['Salary'].head(6)
```

```python
nba['Salary'].sub(5000000).head(6)
```

```python
nba['New Salary'] = nba['Salary'].sub(5000000)
nba.head(6)
```

```python
def new_salary(salary):
    new_salary = 0
    if salary - 5000000 > 0:
        return salary - 5000000
    else:
        return 0
```

```python
nba['Salary'].apply(new_salary).head(4)
```

```python
nba['New Salary'] = nba['Salary'].apply(new_salary)
nba.head(6)
```

```python
nba['Weight'].head(7)
```

```python
nba['Weight'].multiply(0.4535924).head(7)
```

```python
nba['Weight in KG'] = nba['Weight'].multiply(0.4535924)
nba.head(7)
```

```python
def lbs_to_kg(weight_in_kg):
    if weight_in_kg > 0:
        return round(weight_in_kg * 0.4535924, 2)
    else:
        0
```

```python
nba.head(3)
```

```python
nba = pd.read_csv('nba.csv')
nba.head(2)
```

```python
nba['Weight in KG'] = nba['Weight'].apply(lbs_to_kg)
nba.head(3)
```

# A Review of the value_counts Method

```python
nba = pd.read_csv('nba.csv')
nba.head(2)
```

```python
nba.value_counts()
```

```python
nba['Position'].value_counts()
```

```python
nba['Position'].value_counts(normalize=True).mul(100).round(2)
```

```python
nba['Salary'].value_counts().head(3)
```

```python
nba['Salary'].value_counts(normalize=True,ascending=True).multiply(100).round(2).head(15)
```

# Drop Rows with Null Values

```python
nba = pd.read_csv('nba.csv')
nba.tail(2)
```

```python
nba.dropna()
```

```python
nba.tail(2)
```

```python
nba_no_na = nba.dropna()
nba_no_na.tail()
```

```python
nba.tail(2)
```

```python
nba.dropna(how = 'any').tail(2)
```

```python
nba.dropna(how='all').tail(2)
```
