# Rank Values with the rank Method

```python
import pandas as pd
nba = pd.read_csv('nba.csv').dropna(how='all')
nba.head(n=3)
```

```python
nba['Salary'].fillna(0).astype('int')
```

```python
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
```

```python
nba.head(n=3)
```

```python
nba['Salary'].rank()
```

```python
nba['Salary'].rank(ascending=False)
```

```python
nba['Salary Rank'] = nba['Salary'].rank(ascending=False).astype('int')
```

```python
nba.head(n=3)
```

```python
nba.sort_values(by='Salary',ascending=False).head(n=7)
```

```python
nba.sort_values(by='Salary',ascending=False).tail(n=15)
```
