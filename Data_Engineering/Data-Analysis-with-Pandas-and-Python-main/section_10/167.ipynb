# Left Joins

```python
import pandas as pd
```

```python
week_1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week_2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
week_1.head(3)
```

```python
foods.head(3)
```

```python
week_1.merge(foods, how='left')
```

```python
week_1.merge(foods, how='left', on='Food ID')
```

```python
week_1.merge(foods, how='left', on='Food ID', sort=True)
```
