# Merging by Indexes with the left_index and right_index Parameters

```python
import pandas as pd
```

```python
week_1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week_2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
foods.head()
```

```python
customers.head()
```

```python
week_1.merge(customers, how='left', left_on='Customer ID', right_index=True)
```

```python
sales = week_1.merge(customers, how='left', left_on='Customer ID', right_index=True)
```

```python
sales
```

```python
sales.merge(foods, how='left', left_on='Food ID', right_index=True)
```

```python
sales = sales.merge(foods, how='left', left_on='Food ID', right_index=True)
```

```python
sales
```
