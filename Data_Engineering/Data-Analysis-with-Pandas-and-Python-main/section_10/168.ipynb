# The left_on and right_on Parameters

```python
import pandas as pd
```

```python
week_1 = pd.read_csv('Restaurant - Week 1 sales.csv')
week_2 = pd.read_csv('Restaurant - Week 2 sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
week_2.head()
```

```python
customers.head()
```

```python
week_2.merge(customers, how='left', left_on='Customer ID', right_on='ID')
```

```python
week_2.merge(customers, how='left', left_on='Customer ID', right_on='ID').drop('ID', axis='columns')
```

```python
week_2.merge(customers, how='left', left_on='Customer ID', right_on='ID', sort=True).drop('ID', axis='columns')
```
