# Outer joins

```python
import pandas as pd
```

```python
week_1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week_2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
week_1.merge(week_2, how='outer', on='Customer ID')
```

```python
week_1.merge(week_2, how='outer', on='Customer ID', suffixes=[' - Week 1', '  -  Week 2'])
```

```python
week_1.merge(week_2, how='outer', on='Customer ID', suffixes=[' - Week 1', ' - Week 2'], indicator=True)
```

```python
merged_work_week = week_1.merge(week_2, how='outer', on='Customer ID', suffixes=[' - Week 1', ' - Week 2'], indicator=True)
```

```python
merged_work_week['_merge'].value_counts()
```

```python
mask_merge = merged_work_week['_merge'].isin(['left_only', 'right_only'])
merged_work_week[mask_merge]
```
