```python
import pandas as pd
```

```python
week_1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week_2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```
