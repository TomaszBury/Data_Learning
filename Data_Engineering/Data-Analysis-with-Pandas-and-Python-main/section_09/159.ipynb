# The .agg() Method

```python
import pandas as pd
```

```python
fortune = pd.read_csv('fortune1000.csv', index_col='Rank')
sectors = fortune.groupby('Sector')
fortune.head(n=3)
```

```python
sectors['Employees'].mean()
```

```python
sectors.agg()
```

```python
sectors.agg({'Revenue':'min'})
```

```python
sectors.agg({'Profits': 'sum',
                'Employees':'mean'})
```

```python
sectors.agg(['size', 'sum', 'min'])
```

```python
sectors.agg({'Revenue': ['sum','mean'],
            'Profits': 'sum',
            'Employees': 'mean'})
```
