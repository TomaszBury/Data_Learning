# Filering with the where Method

```python
import pandas as pd
james_bond = pd.read_csv('jamesbond.csv', index_col='Film')
james_bond = james_bond.sort_index()
james_bond.head(n=2)
```

```python
james_bond['Actor'] == 'Sean Connery'
```

```python
mask_actor = james_bond['Actor'] == 'Sean Connery'
james_bond[mask_actor]
```

```python
james_bond.where(mask_actor)
```

```python
james_bond.where(mask_actor).dropna()
```

```python
james_bond['Box Office'] > 800
```

```python
mask_box_office = james_bond['Box Office'] > 800
```

```python
james_bond.where(james_bond['Box Office'] > 800)
```

```python
james_bond.where(mask_box_office).dropna()
```

```python
james_bond.where(mask_actor & mask_box_office)
```

```python
james_bond.where(mask_actor & mask_box_office).dropna()
```

```python
james_bond.where(mask_actor | mask_box_office).dropna()
```
