# Retrieve Rows by Index Position with iloc Accessor

```python
import pandas as pd
james_bond = pd.read_csv('jamesbond.csv')
james_bond.head(n=2)
```

```python
james_bond.iloc[0]
```

```python
james_bond.iloc[15]
```

```python
james_bond.iloc[[15,20,0]]
```

```python
james_bond.iloc[0:20:3]
```

```python
100 in james_bond.index
```

```python
james_bond.iloc[100]
```

```python
if 100 in james_bond.index:
    james_bond.iloc[100]
else:
    print('Safety first!')
```
