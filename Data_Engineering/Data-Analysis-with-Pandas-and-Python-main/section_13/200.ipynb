# Creating Pie Charts to Represent proportions

```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bb = data.DataReader(name='BB',data_source='yahoo',start='01-01-2007', end='2022-12-31')
bb.head(n=2)
```

```python
bb['Close'].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 27.3983:
        return 'Above Average'
    else:
        return ' Below Average'
```

```python
plt.style.use('seaborn-poster')
plt.style.use('dark_background')
bb['Close'].apply(rank_performance).value_counts().plot(kind='pie', legend=True)
```
