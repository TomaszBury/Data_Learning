# Modifyuing Plot Asthdtics with matplotlib templates

```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
BB = data.DataReader(name='BB',data_source='yahoo',start='01-01-2007', end='2022-12-31')
BB.head(n=2)
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
```

```python
BB.plot(y='Close')
```

```python
plt.style.use('fast')
```

```python
BB.plot(y='Close')
```

```python
plt.style.use('dark_background')
```

```python
BB.plot(y='Close')
```

```python
plt.style.use('ggplot')
```

```python
BB.plot(y='Close')
```

```python
plt.style.use('seaborn-poster')
BB.plot(y='Close')
```
