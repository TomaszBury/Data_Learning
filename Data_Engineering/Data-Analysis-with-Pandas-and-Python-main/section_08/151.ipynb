# The pivot_table() Method

```python
import pandas as pd
```

```python
pd.read_csv('foods.csv')
```

```python
foods = pd.read_csv('foods.csv')
```

```python
foods.head(n=2)
```

```python
foods.pivot_table(values='Spend', index='Gender', aggfunc='mean')
```

```python
foods.pivot_table(values='Spend', index='Gender', aggfunc='sum')
```

```python
foods.pivot_table(values='Spend', index='Item', aggfunc='mean')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], aggfunc='mean')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns='City', aggfunc='mean')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns=['Frequency', 'City'], aggfunc='mean')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], columns=['City'], aggfunc='mean')
```

```python
foods.pivot_table(values='Spend', index='Item', aggfunc='count')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], aggfunc='max')
```

```python
foods.pivot_table(values='Spend', index=['Gender', 'Item'], aggfunc='min')
```

```python
foods.pivot_table(values='Spend', index=['Item'], aggfunc='max')
```

```python
foods.pivot_table(values='Spend',index='City', aggfunc='mean')
```

```python
pd.pivot_table(data=foods, values='Spend',index='City', aggfunc='mean')
```

```python
pd.pivot_table(data=foods, values='Spend', index='City', aggfunc='max')
```
