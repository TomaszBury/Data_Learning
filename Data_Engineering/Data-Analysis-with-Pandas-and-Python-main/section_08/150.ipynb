# The pivot Method

```python
import pandas as pd
```

```python
pd.read_csv('salesmen.csv')['Salesman'].value_counts()
```

```python
salesmen = pd.read_csv('salesmen.csv', parse_dates=['Date'])
```

```python
salesmen.head(n=2)
```

```python
len(salesmen)
```

```python
salesmen['Salesman'].value_counts()
```

```python
salesmen = salesmen.astype('category')
```

```python
salesmen.info()
```

```python
salesmen.head(n=12)
```

```python
salesmen.pivot()
```

```python
salesmen.pivot(columns=['Salesman'])
```

```python
salesmen.pivot(index='Date', columns='Salesman', values='Revenue')
```

```python
salesmen.pivot(index='Date', columns='Salesman')
```

```python
salesmen = salesmen.pivot(index='Date', columns='Salesman', values='Revenue')
```
