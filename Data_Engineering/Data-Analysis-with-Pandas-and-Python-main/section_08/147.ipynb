# The .unstack() Method, Part 1

```python
import pandas as pd
```

```python
world_stats = pd.read_csv('worldstats.csv', index_col=['country', 'year'])
world_stats.head(n=3)
```

```python
world_stats.stack()
```

```python
world_stats = world_stats.stack()
world_stats.head(n=3)
```

```python
world_stats.unstack()
```

```python
world_stats.unstack().unstack()
```

```python
world_stats.unstack().unstack().unstack()
```

```python
world_stats.unstack().unstack().unstack().to_frame()
```
